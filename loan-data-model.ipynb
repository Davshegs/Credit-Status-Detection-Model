A Model To Predict the Loan Status

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Loading the dataset
loan_df = pd.read_csv('loan-data-preprocessing.csv')
```

```python
# Seperating the dataset for the model since we need only the amount in EUR
loan_model = loan_df.iloc[:, [2, 4, 5, 7, 9, 11, 12, 13, 14, 15, 16]]
```

```python
loan_model
```

```python
# spliting the data into train and test data

train_data, test_data = train_test_split(loan_df, test_size = 0.2, random_state = 42)
```

```python
# Creating an instant of Logistic Regression
logreg = LogisticRegression()
```

```python
# Seperating the target variable ('survived') fron the rest of the trianing data
x_train = train_data.drop('loan_status', axis = 1)
y_train = train_data['loan_status']
```

```python
# Training the data with Logistic Regression model

logreg.fit(x_train, y_train)
```

```python
# Seperating the  target varable and the rest of the testing data
x_test = test_data.drop('loan_status', axis = 1)
y_test = test_data['loan_status']
```

```python
# Predicting with the Logistic Regression model
prediction = logreg.predict(x_test)
```

```python
# Classification Report
print('Classification Report:\n ', classification_report(y_test, prediction))
```

```python
# Confussion Matrix
print('Confusion Matrix:\n ', confusion_matrix(y_test, prediction))
```

```python
# Accuracy Score
print('Accuracy Score:\n ', accuracy_score(y_test, prediction))
```
